```python
import pandas as pd

data=pd.read_csv('dataset.csv')

df = pd.DataFrame(data)

# Create a dictionary that maps states to regions
state_to_region = {
    'Andhra Pradesh':'South',
'Arunachal Pradesh':'East',
'Assam':'East',
'Bihar':'North',
'Chhattisgarh':'North',
'Goa':'South',
'Gujarat':'West',
'Haryana':'North',
'Himachal Pradesh':'North',
'Jharkhand':'North',
'Karnataka':'South',
'Kerala':'South',
'Madhya Pradesh':'North',
'Maharashtra':'South',
'Manipur':'East',
'Meghalaya':'East',
'Mizoram':'East',
'Nagaland':'East',
'Odisha':'South',
'Punjab':'North'
}

# Create a new column called 'Region' by mapping the values in the 'State' column
df['Region'] = df['State'].map(state_to_region)

display(df)

# Group the DataFrame by the 'Region' column and calculate the mean population for each region
region_populations = df.groupby('Region')['Population (millions)'].mean()

# Display the resulting Series
display(region_populations)
```

## The dtype parameter is used when creating arrays in NumPy to specify the data type of the elements in the array. NumPy arrays are homogeneous, meaning that all elements in the array must have the same data type. The dtype parameter allows you to control the data type of the elements when creating an array.

```python
import numpy as np

# Create an array of integers
arr_int = np.array([1, 2, 3], dtype=int)
print(f'Array of integers: {arr_int}')

# Create an array of floating-point numbers
arr_float = np.array([1, 2, 3], dtype=float)
print(f'Array of floating-point numbers: {arr_float}')

# Create an array of complex numbers
arr_complex = np.array([1, 2, 3], dtype=complex)
print(f'Array of complex numbers: {arr_complex}')
```

The second array is created with dtype=float, which means that its elements are floating-point numbers. The third array is created with dtype=complex, which means that its elements are complex numbers.

# Specifying the data type using the dtype parameter can be useful in several situations. For example, it can help you control the amount of memory used by the array, since different data types have different memory requirements. It can also help you ensure that the calculations performed on the array are done using the desired level of precision.

## Heres a Python program that prints numbers from 1 to 100, but for multiples of 3, prints Fizz instead of the number, and for multiples of 5, prints Buzz. For numbers that are multiples of both 3 and 5, it prints FizzBuzz:

```python
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

## This code uses a for loop to iterate over the numbers from 1 to 100. For each number, it checks if the number is divisible by both 3 and 5 using the modulo operator (%). If the number is divisible by both 3 and 5, it prints FizzBuzz. If the number is only divisible by 3, it prints Fizz. If the number is only divisible by 5, it prints Buzz. If the number is not divisible by either 3 or 5, it prints the number itself.

<h1 style="background-color:tomato;">29.Palindrome Check: Write a function to check if a given string is a palindrome (reads the same forwards and backwards).</h1>

## Sure! Here's a Python function that checks if a given string is a palindrome:



```python
def is_palindrome(s):
    return s == s[::-1]
```

## This function takes a string s as an input and returns True if the string is a palindrome, and False otherwise. The function checks if the string is equal to its reverse by using the slice notation s[::-1] to reverse the string.

# Heres an example of how you can use this function:

```python
print(is_palindrome("racecar")) # True
print(is_palindrome("hello"))   # False
```

## In this example, we call the is_palindrome function with two different strings as inputs. The first call checks if the string racecar is a palindrome, and the function returns True because racecar reads the same forwards and backwards. The second call checks if the string hello is a palindrome, and the function returns False because hello does not read the same forwards and backwards.

<h1 style="background-color:tomato;">32.Write a function to count the number of words in a given sentence.</h1>

## Heres a Python function that counts the number of words in a given sentence:

```python
def count_words(sentence):
    words = sentence.split()
    return len(words)
```

## This function takes a sentence as an input and returns the number of words in the sentence. The function uses the split method to split the sentence into a list of words. The split method splits the sentence at each occurrence of a whitespace character (such as a space, tab, or newline) and returns a list of the resulting substrings. The function then uses the len function to count the number of elements in the list of words, which gives us the number of words in the sentence.

# Heres an example of how you can use this function:

```python
sentence = "This is an example sentence."
num_words = count_words(sentence)
print(f"The sentence has {num_words} words.")
```

## In this example, we call the count_words function with a sentence as an input. The function returns the number of words in the sentence. When you run this code, it will print the resulting count: The sentence has 5 words.

### Heres an example of how you can use the genfromtxt function to read data from a CSV file into a NumPy array:

```python
import numpy as np

data = np.genfromtxt('emp.csv', delimiter=',')

```

## In this example, we use the genfromtxt function to read data from a file called data.csv. The delimiter parameter is used to specify that the values in the file are separated by commas. The resulting data array contains the data from the CSV file.
