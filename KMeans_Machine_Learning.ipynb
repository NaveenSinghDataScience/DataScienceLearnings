# Here's an example of how to perform K-means clustering in Python using the popular machine learning library, scikit-learn. In this example, we'll generate some random data and then apply K-means clustering to it.

# Now, let's create a Python script for K-means clustering:

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate random data for clustering
np.random.seed(0)
X = np.random.rand(100, 2)

# Define the number of clusters (K)
k = 3

# Create a K-means model
kmeans = KMeans(n_clusters=k)

# Fit the model to the data
kmeans.fit(X)

# Get cluster labels for each data point
labels = kmeans.labels_

# Get cluster centers
centers = kmeans.cluster_centers_

# Plot the data points and cluster centers
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, label='Cluster Centers')
plt.legend()
plt.title('K-means Clustering')
plt.show()
```

# In this example:

 * We import the necessary libraries: NumPy for data generation, scikit-learn for K-means clustering, and Matplotlib for visualization.
 * We generate random data (X) for clustering. You would replace this with your own dataset in a real-world scenario.
 * We specify the number of clusters (k) to be created by K-means.
 * We create a K-means model and fit it to the data.
 * We get the cluster labels for each data point and the cluster centers.
### Finally, we visualize the data points with colors representing their clusters and mark the cluster centers in red.
