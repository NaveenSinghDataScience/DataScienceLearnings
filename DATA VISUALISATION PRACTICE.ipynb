```python
import seaborn as cb
import numpy as np
import matplotlib.pyplot as mpl

months=['jan','feb','mar','apr','may']
tempreature=np.array([
    [22,35,23],
    [45,12,43],
    [12,76,33],
    [21,98,23],
    [75,12,65]
])

cb.heatmap(tempreature,xticklabels=['Morning','Noon','Night'],yticklabels=months)

mpl.xlabel('TIME OF THE DAY')
mpl.ylabel('MONTHS')

mpl.show()
```
